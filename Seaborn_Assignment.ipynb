Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular data visualization library in Python that provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots you can create using Seaborn and their common use cases:

Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a data point, making it useful for identifying patterns, clusters, or trends in the data.
Seaborn Function: sns.scatterplot()

Histogram:
Use: Histograms are used to display the distribution of a single continuous variable by dividing it into bins and counting the frequency of data points in each bin. It helps in understanding the data's central tendency and spread.
Seaborn Function: sns.histplot()

Bar Plot:
Use: Bar plots are useful for visualizing the relationship between a categorical variable and a numerical variable. They can show comparisons between different categories or groups.
Seaborn Function: sns.barplot()

Box Plot (Box-and-Whisker Plot):
Use: Box plots are used to visualize the distribution and spread of a numerical variable, including the presence of outliers. They are especially useful for comparing the distribution of a variable across different categories or groups.
Seaborn Function: sns.boxplot()

Heatmap:
Use: Heatmaps are used to represent data in a 2D grid format, where each cell's color intensity corresponds to the value of a variable. They are commonly used to visualize correlations between variables in a dataset.
Seaborn Function: sns.heatmap()

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
plt.figure(figsize=(6, 3))
sns.lineplot(x="timepoint", y="signal", data=fmri)
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix for the heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))  # Set the figure size
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)

# Customize the plot
plt.title("Passenger Count by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
