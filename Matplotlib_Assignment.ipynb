Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
'''Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much 
one variable is affected by another.'''

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.figure(figsize=(6,2))
plt.scatter(x,y)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("X vs Y Scatter")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
#subplot() function adds subplot to a current figure at the specified grid position.
```

```python
# Data for four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 6))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 2', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 1')
plt.legend()
# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 2')
plt.legend()
# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 3')
plt.legend()

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='purple')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 4')
plt.legend()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
#Bar plot shows the relationship between a numeric and a categoric variable.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(15,8))

plt.subplot(2, 2, 1)
plt.bar(company, profit, color = 'r')
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company vs Profit")
#Horizontal Plot
plt.subplot(2, 2, 2)
plt.barh(company, profit, color = 'purple')
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company vs Profit")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
'''A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution 
of a dataset. It displays a summary of the data's central tendency (median), 
spread (interquartile range), and potential outliers. Box plots are particularly useful 
for comparing the distribution of multiple datasets or identifying the presence of outliers in a dataset.'''

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots()

# Create the box plot
box_data = [box1, box2]
labels = ['Box 1', 'Box 2']
ax.boxplot(box_data, labels=labels)

# Set plot title and labels
ax.set_title('Box Plot Example')
ax.set_xlabel('Box')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
