App Routing means mapping the URLs to a specific function that will handle the logic for that URL. We use app routing to route the specific url to the respective function.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
# CODE
from flask import Flask

app = Flask(__name__)

@app.route("/")
def welcome_company():

    details = '''<h1>Company Name: ABC Corporation<br>
                 Location: India<br>
                 Contact Detail: 999-999-9999<h1>'''
    return details

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation<h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
#url_for() function is used for URL building

from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def welcome_company():

    details = '''<h1>Company Name: ABC Corporation<br>
                 Location: India<br>
                 Contact Detail: 999-999-9999<h1>'''
    return details

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation<h1>"

if __name__=="__main__":
    with app.test_request_context():
        print("URL for 'welcome' route:", url_for('welcome'))
        print("URL for 'welcome_company' route:", url_for('welcome_company'))
        app.run(host="0.0.0.0")
```
